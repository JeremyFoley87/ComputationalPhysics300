From Electromagnetic class we went over the concept of electromagnetic waves, which concists of electric field and the magnetic feild.  This two fields are perpendicular to each other and constantlly oscillating.  You can find the general propogation of the of the Electromagnetic wave by finding the cross product of the E and B vectors.  

The general electrical field component in this case is E(r,t) = Esin((wt)(sin(kz) + cos(kx))

This is from the specific equation on page 196 in Classical Electromagnetic Radiation by Marion
                                                                     

```python
import numpy
import matplotlib.pyplot as plt
from math import pi 
from numpy import sin
from numpy import cos


#Chosing values and constants
plt.figure(figsize = (10,6))

t = numpy.arange(0,10,0.1)
z = numpy.arange(0,10,0.1)
E = 5
k = 2 * pi
w = 4 * pi

x = E*sin(w*t)*(sin(k*z) + cos(k*z))



plt.title("Electric field component")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t,x)
plt.show()

Fourier = numpy.fft.fft(x)

plt.figure(figsize = (10,6))
plt.plot(abs(Fourier))
plt.title("Fourier Series")

           

```
