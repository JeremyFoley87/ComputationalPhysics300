```python
## Jeremy Foley. Homework #2 ##

#Imported libraries that will be used
import numpy as np
import matplotlib.pyplot as plt
from math import log, exp

#Multiplicity calculation from example
def factorial (N):
    N = int(N)
    total = 1
    if N < 0:
        print("Factorial error: N must be greater than 0")
    elif N > 1:
        for i in range(2, N+1):
            total *= i
    return total

#Found this multiplicity in the example as a better way to compute the multiplicity
def Multiplicity(N,n):
    return exp(log(factorial(N+n-1)) - log(factorial(n)) - log(factorial(N-1))) 

def Plot_Omega(Na, Nb, q):    

#Plotting the multiplicities of the variables
#Starting at q + 1 because the system can't start at 0
    qa_list = range(q + 1)
    Omega_A = []
    Omega_B = []
   
    for qa in qa_list:
        Omega_A.append(Multiplicity(Na, qa))
        Omega_B.append(Multiplicity(Nb, (q - qa))) 
#Defining Omega        
    Omega_A = numpy.asarray(Omega_A)
    Omega_B = numpy.asarray(Omega_B)
    Omega_T = Omega_A * Omega_B
#Labling plot with width, vs, multiplicity, and color

    plt.plot(qa_list, Omega_T, linewidth=1,)
    plt.title("$q_a$ vs. Total Multiplicity")
    plt.xlabel("$q_a$")
    plt.ylabel("Total Multiplicity") 
    plt.plot(qa_list, Omega_A, '-g')
    plt.plot(qa_list, Omega_B, '-r')
```

```python
Plot_Omega(300, 200, 100)
```
