```python
## Jeremy Foley HW #9 Part 1 ##

import matplotlib.pyplot as plt
import numpy as np
from pylab import *
import sys


data = np.loadtxt("http://www-personal.umich.edu/~mejn/cp/data/circular.txt",float)
print(np.shape(data))

fdata = np.fft.fft2(data) 
#absolute value to get rid of complex numbers and < 100 to get rid of small coefficents##
fdata[np.abs(fdata)<100] = 0 
#getting back the information through inverse fft
idata = np.fft.ifft2(fdata)

##I was trying to show the file size using A.nbytes and sys.getsizeoff##
A = array(data)
A.nbytes
print(A.nbytes)

Af = array(np.real(fdata))
Af.nbytes
print(Af.nbytes)

sys.getsizeof([data])
print(sys.getsizeof([data]))
sys.getsizeof([idata])
print(sys.getsizeof([idata]))

plt.title('Original Data')
plt.imshow(data)
plt.colorbar()
plt.show()

plt.title('Compressed Data')
plt.imshow(np.real(idata))
plt.colorbar()
plt.show()
```
