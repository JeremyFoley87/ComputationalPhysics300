```python
#Jeremy Foley Homework #3 #

#imported libraries needed#

import numpy as np
import matplotlib.pyplot as plt

#definition of the mandelbrot set from lecture#

def mandelbrot(Re, Im, Max):
    c = complex(Re, Im)
    z = 0.0j
#for loop needed for mandelbrot set where abs value of z must be >= 2    
    for i in range(Max):
        z = z*z + c
        
        if(z.real*z.real + z.imag*z.imag) >= 2:
            return i
        
    return Max

##image resolution##

columns = 1000
rows = 1000

#numpy.zero = Return a new array of given shape and type, with zeros.#
result = np.zeros([rows, columns])

#row_index defines the real axis#
#colum_index defines the imag axis#
#Enumerate() method adds a counter to an iterable and returns it in a form of enumerate object#
for row_index, Re in enumerate(np.linspace(-2,2,num=rows)):
    for column_index, Im in enumerate(np.linspace(-2,2,num=columns)):
        result[row_index, column_index] = mandelbrot(Re, Im, 200)
        
#plotting information#
plt.figure(dpi=300)
plt.imshow(result.T, cmap= 'gist_heat', interpolation='bilinear', extent=[-2,2,-2,2])
plt.xlabel('Real axis')
plt.ylabel('Imaginary axis')

plt.show()

        
```
