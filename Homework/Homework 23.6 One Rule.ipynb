```python
#Download datasets using scikit-learn

from sklearn.datasets import load_iris
dataset = load_iris()
X = dataset.data
Y = dataset.target
#Columns are: Sepal Lenght, Sepal Width, Petal Length, Petal Width

print(X)
print(Y)
```

```python
#This plot shows the original data before implementing One Rule Algorithm
from collections import OrderedDict
import pandas as pd

Iris_data = OrderedDict({'Sepal Length ': X[:,0],
             'Sepal Width': X[:,1],
             'Petal Length': X[:,2],
             'Petal Width': X[:,3],
             'Label': Y})

df = pd.DataFrame(Iris_data)
df
```

```python
#The way shown in class was giving me some kind of bound error so I decided to reshape them
import numpy as np

X_0 = np.array( X[:,0] > X[:,0].mean(), dtype = 'int')
X_1 = np.array( X[:,1] > X[:,1].mean(), dtype = 'int')
X_2 = np.array( X[:,2] > X[:,2].mean(), dtype = 'int')
X_3 = np.array( X[:,3] > X[:,3].mean(), dtype = 'int')

X_d0 = X_0.reshape(150,1)
X_d1 = X_1.reshape(150,1)
X_d2 = X_2.reshape(150,1)
X_d3 = X_3.reshape(150,1)
Y_d = Y.reshape(150,1)
```

```python
#Testing to see if the reshape created the same data frame
Iris_data_d = OrderedDict({'Sepal Length ': X_d0[:,0],
             'Sepal Width': X_d1[:,0],
             'Petal Length': X_d2[:,0],
             'Petal Width': X_d3[:,0],
             'Label': Y})

df = pd.DataFrame(Iris_data_d)
df
```

```python
X_final = np.hstack([X_d0,X_d1,X_d2,X_d3,Y_d])
```

```python
values = []
a = ['Sepal length','Sepal width','Petal length','Petal width']
b = ['Setosa','Versicolor','Virginica']
c = '%'
for i in range(4):
    for j in range(3):
        selection = np.logical_and(X_final[:,i]>.9,X_final[:,4] == j)
        values.append((a[i],b[j],'selected values:', len(X_final[selection]), 'percent', ((len(X_final[selection])/len(X_final[:,i]))*100)))
```
