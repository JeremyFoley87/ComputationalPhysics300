```python
## Jeremy Foley HW#5 ##

from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt

def Debye(T):
    integrand = lambda x: (x**4 * (np.exp(x))/((np.exp(x))-1)**2)
    
    return 9*V*P*B*(T / D)**3*(integrate.quad(integrand, 0 , D / T)[0]) ##[0] give the first value##

V= 0.001       ##Volume of the solid##
P= (6.022e28)  ##Density of atoms##
B= (1.38e-23)  ##Boltzmann constant##
D= 428         ##Debye temperature##


Cv_array = []
Temps = range(5, 500) ##Range needed for HW##

for T in Temps:
    Cv_array.append(Debye(T))

plt.title("Cv as a function of temperature")
plt.xlabel("Temp")
plt.ylabel("Cv")
plt.plot(Temps,Cv_array)
plt.show()

Debye(200) ##Used to check with another student if my function worked## 
```
